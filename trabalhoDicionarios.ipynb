```python
import requests

url_poke = "https://pokeapi.co/api/v2/pokemon/ditto"
info = requests.get(url_poke).json()

habilidades = []
for a in info['abilities']:
    nome_hab = a['ability']['name']
    habilidades.append(nome_hab)

print("Habilidades do Ditto:", habilidades)
print("------------------------------")

url_hab = "https://pokeapi.co/api/v2/ability/battle-armor"
dados = requests.get(url_hab).json()

texto_efeito = ""
for e in dados['effect_entries']:
    if e['language']['name'] == 'en':
        texto_efeito = e['short_effect']

quem_tem = []
for p in dados['pokemon']:
    pokenome = p['pokemon']['name']
    quem_tem.append(pokenome)

print("Efeito da habilidade:", texto_efeito)
print("Pokémons que possuem:", quem_tem)
```

```python
import time

id_dep = 204534 
ano = 2023

url = "https://dadosabertos.camara.leg.br/api/v2/deputados/" + str(id_dep) + "/despesas"

filtro = {'ano': ano, 'itens': 100}

info = requests.get(url, filtro)
dados = info.json()

links = dados['links']
ultima_p = 1

for l in links:
    if l['rel'] == 'last':
        ultimo = l['href']
        parte1 = ultimo.split('pagina=')[1]
        num = parte1.split('&')[0]
        ultima_p = int(num)

quantia = {}
nomes = {}

for p in range(1, ultima_p + 1):
    filtro['pagina'] = p
    
    res = requests.get(url, filtro)
    
    if res.status_code == 200:
        conteudo = res.json()
        gastos = conteudo['dados']
        
        for g in gastos:
            cnpj = g['cnpjCpfFornecedor']
            valor = g['valorLiquido']
            fornecedor = g['nomeFornecedor']
            
            if cnpj in quantia:
                quantia[cnpj] = quantia[cnpj] + valor
            else:
                quantia[cnpj] = valor
                nomes[cnpj] = fornecedor
    
    time.sleep(1)

def ordena(item):
    return item[1]

lista = list(quantia.items())
ranking = sorted(lista, key=ordena, reverse=True)

print("------------------------------------------------------------")
print("FORNECEDORES (Ano:", ano, ")")
print("------------------------------------------------------------")

for item in ranking[:10]:
    cnpj_venda = item[0]
    total_venda = item[1]
    nome_fornecedor = nomes[cnpj_venda]
    
    print("Fornecedor:", nome_fornecedor)
    print("CNPJ:", cnpj_venda, "| Total: R$", round(total_venda, 2))
    print("")
```

```python
import requests
import time
import nltk
import string
from nltk.corpus import stopwords

nltk.download('stopwords')
nltk.download('punkt')

id_dep = 204534 
data_ini = '2023-01-01'

url = "https://dadosabertos.camara.leg.br/api/v2/deputados/" + str(id_dep) + "/discursos"

filtro = {'dataInicio': data_ini, 'itens': 100}

info = requests.get(url, filtro)
dados = info.json()

links = dados['links']
ultima_p = 1

for l in links:
    if l['rel'] == 'last':
        ultimo = l['href']
        parte1 = ultimo.split('pagina=')[1]
        num = parte1.split('&')[0]
        ultima_p = int(num)

texto = ""

for p in range(1, ultima_p + 1):
    filtro['pagina'] = p
    res = requests.get(url, filtro)
    
    if res.status_code == 200:
        conteudo = res.json()
        falas = conteudo['dados']
        
        for f in falas:
            if f['transcricao']:
                texto = texto + " " + f['transcricao']
    
    time.sleep(1)

texto = texto.lower()
vazias = set(stopwords.words('portuguese'))
todas_palavras = nltk.word_tokenize(texto)

contagem = {}

for p in todas_palavras:
    if p not in vazias and p not in string.punctuation and len(p) > 2:
        if p in contagem:
            contagem[p] = contagem[p] + 1
        else:
            contagem[p] = 1

def ordena(item):
    return item[1]

lista = list(contagem.items())
ranking = sorted(lista, key=ordena, reverse=True)

print("--------------------------------------------------")
print("TOP 30 PALAVRAS - DEPUTADA ID:", id_dep)
print("--------------------------------------------------")

for item in ranking[:30]:
    palavra = item[0]
    total = item[1]
    print("Palavra:", palavra, "| Repetida:", total, "vezes")
```
